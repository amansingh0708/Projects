```python
#Q5.
#synthetic correlated dataset
np.random.seed(3)
size = 200
x1 = np.random.normal(0, 1, size)
x2 = x1 + np.random.normal(0, 0.2, size)
x3 = x1 - x2 + np.random.normal(0, 0.3, size)
x4 = np.random.normal(0, 1, size)
#DataFrame
df = pd.DataFrame({'Feature 1': x1, 'Feature 2': x2, 'Feature 3': x3, 'Feature 4': x4})
#correlation matrix
corr = df.corr()
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#Q4.
#Data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
```

```python
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.update_layout(title='3D Surface Plot: z = sin(sqrt(x² + y²))',
                  scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Z'))
fig.show()
```

```python
#Q2.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
output_notebook()
N = 200
x = np.random.rand(N) * 100
y = np.random.rand(N) * 100
sizes = np.random.rand(N) * 30 + 10
colors = np.random.choice(["red", "green", "blue", "purple", "orange"], N)
source = ColumnDataSource(data=dict(x=x, y=y, size=sizes, color=colors))
p = figure(title="Random Scatter with Sizes and Colors")
p.scatter(x='x', y='y', size='size', color='color', alpha=0.6, source=source)
show(p)
```

```python
#Q5.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.transform import transform
from bokeh.palettes import Viridis256
import numpy as np
import pandas as pd
output_notebook()
#Data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
#Dataframe
df = pd.DataFrame({
    'x': xx.flatten(),
    'y': yy.flatten(),
    'value': data_heatmap.flatten()
})
```

```python
mapper = LinearColorMapper(palette=Viridis256, low=df.value.min(), high=df.value.max())
p = figure(title="Bokeh Heatmap", x_axis_label="X", y_axis_label="Y",
           x_range=(0, 1), y_range=(0, 1), toolbar_location=None)

p.rect(x='x', y='y', width=0.1, height=0.1, source=df,
       fill_color=transform('value', mapper), line_color=None)
color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')
show(p)
```
